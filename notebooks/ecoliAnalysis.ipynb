```python
import os

# Directory containing your FASTA files
fasta_directory = 'ecoli/'

# Get a list of all FASTA files in the directory
fasta_files = [filename for filename in os.listdir(fasta_directory) if filename.endswith('.fna')]

# Create a dictionary to store file contents
file_contents = {}

# Iterate over each file
for filename in fasta_files:
    combinedLine = ""
    
    with open(os.path.join(fasta_directory, filename), 'r') as file:
        # Skip the first line in each file
        next(file)
        
        for line in file:
            combinedLine += line.strip()

    file_contents[filename] = combinedLine  # Store the content in the dictionary

# Find the length of the shortest file content
min_length = min(len(content) for content in file_contents.values())

i = 0
while i < min_length:
    # Get the character at the current position for all files
    characters = [content[i] for content in file_contents.values()]

    # Check if all characters at the current position are the same
    if all(char == characters[0] for char in characters):
        print("Commonality detected at Position " + str(i+1))
        print("Common letter:", characters[0])
    
    i = i + 1
```
